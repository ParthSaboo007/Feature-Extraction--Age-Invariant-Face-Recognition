```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import norm

data1 = df3['Predicted_score_scaled']
data2 = df2['Predicted_score_scaled']

mu1, std1 = norm.fit(data1)
mu2, std2 = norm.fit(data2)  

plt.hist(data1, bins=100, density=True, alpha=0.6, color='y')
x1min, x1max = plt.xlim()
x1 = np.linspace(xmin, xmax, 100)
p1 = norm.pdf(x1, mu1, std1)
plt.plot(x1, p1, 'k', linewidth=2,color='g')


plt.hist(data2, bins=100, density=True, alpha=0.6, color='b')
x2min, x2max = plt.xlim()
x2 = np.linspace(xmin, xmax, 100)
p2 = norm.pdf(x2, mu2, std2)
plt.plot(x2, p2, 'k', linewidth=2,color='r')


def solve(m1,m2,std1,std2):
  a = 1/(2*std1**2) - 1/(2*std2**2)
  b = m2/(std2**2) - m1/(std1**2)
  c = m1**2 /(2*std1**2) - m2**2 / (2*std2**2) - np.log(std2/std1)
  return np.roots([a,b,c])

result = solve(mu1,mu2,std1,std2)

print(result)
plt.show()
```

```python

fp_count = 0
tp_count = 0
fn_count = 0
tn_count = 0

for ind in pf.index:
  if pf['actual_score'][ind]==1 and pf['th1_score'][ind]==1:
       tp_count=tp_count+1
  elif pf['actual_score'][ind]==1 and pf['th1_score'][ind]==0:
    fn_count=fn_count+1
  elif pf['actual_score'][ind]==0 and pf['th1_score'][ind]==1:
    fp_count=fp_count+1
  else:
    tn_count=tn_count+1

total_matches = pf['th1_score'].count()
print(fp_count+tp_count+fn_count+tn_count, pf['th1_score'].count())
crr = tp_count/total_matches
far = fp_count/total_matches
ffr = fn_count/total_matches
acc = 100-(far+ffr)/2

print('far=',far)
print('ffr',ffr)
```
